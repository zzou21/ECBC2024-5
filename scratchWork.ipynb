```python
with open(jsonOCR, "r") as file:
    storage = json.load(file)
wordCounter = 0
listOfGovernors = ["Governor", "governor", "gouernor"]
for checkWord in listOfGovernors:

    for page, content in storage.items():
        if checkWord in content:
            governorIndex = content.index(checkWord)
            print(f"Found in page {page}. Context: {content[governorIndex - 100:governorIndex], content[governorIndex:governorIndex + 100]} \n\n\n")
            wordCounter += 1

print(f"Governor appeared in total {listOfGovernors} times.")
```
